1. What is hypothesis testing in statistics?
 - Hypothesis testing in statistics is a method used to make decisions or inferences about population parameters based on sample data. It helps determine whether the observed data provides enough evidence to reject a specific assumption (called the null hypothesis).
2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
 - Null Hypothesis (H₀):
    - The null hypothesis is a statement of no effect, no difference, or status quo.
    - It is the hypothesis that you're trying to test against.
    - It assumes that any observed difference is due to chance or random variation.
 - Example:
    - In a drug trial:
    - H₀: The new drug has no effect on blood pressure (i.e., the mean change = 0).
 - Alternative Hypothesis (H₁ or Ha):
    - The alternative hypothesis is what you hope to show or support.
    - It suggests that there is a real effect, difference, or relationship.
    - It’s the opposite of the null hypothesis.
  - Same example:
    - Ha: The new drug does affect blood pressure (i.e., the mean change ≠ 0).

3. What is the significance level in hypothesis testing, and why is it important?
 - The significance level in hypothesis testing, often denoted by α (alpha), is a threshold that helps you decide whether to reject the null hypothesis.
 - Why Is It Important?
    - 1. Decision Rule:
    - You compare the p-value to α to decide
    - If p-value ≤ α → Reject H₀
    - If p-value > α → Fail to reject H₀
 - Controls the Risk of False Alarms:
    - Helps keep the chance of wrongly declaring a significant effect under control.
 - Balances Risk and Sensitivity:
    - Lower α → fewer false positives, but risk missing real effects (higher Type II error).
    - Higher α → more sensitive to detecting effects, but more risk of false positives.

4. What does a P-value represent in hypothesis testing?
 - The p-value is the probability of observing your sample results, or something more extreme, assuming the null hypothesis (H₀) is true.
 - In simpler terms:
 - It tells you how surprising your data is if there really is no effect.
 - Key Points:
    - A small p-value means your data is unlikely under H₀ → may be evidence against H₀.
    - A large p-value means your data is plausible under H₀ → not enough evidence to reject H₀.
5. How do you interpret the P-value in hypothesis testing?
 - The p-value tells you how likely it is to see your sample results (or something more extreme) if the null hypothesis (H₀) is true.

6. What are Type 1 and Type 2 errors in hypothesis testing?
 - Type I Error (False Positive):
    - You reject the null hypothesis even though it’s actually true.
    - You think you found an effect, but it’s really just random chance.
    - Probability of this error = α (significance level).
 - Example: You say a drug works, but it doesn’t.
    - “You got fooled by randomness.”
 - Type II Error (False Negative):
    - You fail to reject the null even though it’s false.
    - There is a real effect, but your test missed it.
    - Probability of this error = β.
    - The power of a test = 1 − β (i.e., how likely it is to detect a true effect).
  - Example: You say a drug doesn’t work, but it actually does.
    -  “You missed something real.”
7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
 - 1. Two-Tailed Test (Non-Directional)
    - Purpose: Tests for the possibility of an effect in both directions (greater than or less than the hypothesized value).
    - Null Hypothesis (H₀): No effect or difference (e.g., μ = μ₀).
    - Alternative Hypothesis (H₁): There is an effect, but direction is not specified (e.g., μ ≠ μ₀).
    - Critical Region: Split into both tails of the distribution (e.g., both extreme positive and negative values).
    - When to Use: When you want to detect any significant difference, regardless of direction.

 - Example: Testing if a new drug has a different effect (either higher or lower) than a placebo.
 - 2. One-Tailed Test (Directional)
    - Purpose: Tests for an effect in only one direction (either greater than or less than the hypothesized value).
    - Null Hypothesis (H₀): No effect or difference in the specified direction (e.g., μ ≤ μ₀ or μ ≥ μ₀).
    - Alternative Hypothesis (H₁): There is an effect in one specific direction (e.g., μ > μ₀ or μ < μ₀).
    - Critical Region: Located in only one tail of the distribution.
    - When to Use: When you have a strong theoretical basis to expect an effect in a specific direction.
 - Example: Testing if a new drug is more effective than an existing one (not just different).

8. What is the Z-test, and when is it used in hypothesis testing?
 - The Z-test is a statistical hypothesis test used to determine whether there is a significant difference between sample data and a population parameter (like the mean) when the population standard deviation (σ) is known, and the sample size is sufficiently large (typically n ≥ 30). It is based on the Z-distribution (standard normal distribution).
 - When to Use the Z-Test:
    - The Z-test is appropriate under the following conditions:
    - 1. Known Population Standard Deviation (σ): The test requires prior knowledge of σ.
    - 2. Large Sample Size (n ≥ 30): Due to the Central Limit Theorem, the sampling distribution approximates normality even if the underlying population isn’t normal.
    - 3. Data is Continuous: Typically used for means, proportions (with some adjustments), or differences between means.

10. What is the T-distribution, and when should it be used instead of the normal distribution?
 - The T-distribution (also called Student's t-distribution) is a bell-shaped, symmetric distribution like the normal distribution — but with heavier tails (more variability).
     - It’s used when you’re estimating population parameters and don’t know the population standard deviation.
  - When to Use the T-Distribution Instead of Normal (Z):
     - Use the T-distribution when:
     - The population standard deviation (σ) is unknown, and
     - You are using the sample standard deviation (s) instead, and
     - The sample size is small (typically n < 30)

20. What is an ANOVA test, and what are its assumptions?
 - The ANOVA test — short for Analysis of Variance — is a powerful statistical method used when you want to compare the means of three or more groups. Instead of running multiple t-tests (which increases error), ANOVA lets you test all groups at once.
- Key Assumptions of ANOVA:
 - Independence
    - Each observation is independent of the others (no overlapping data)
 - Normality
    - The data within each group should be approximately normally distributed
    - ANOVA is fairly robust to mild violations if sample sizes are similar
 - Homogeneity of Variance (Homoscedasticity)
    - All groups should have similar variances
    - You can test this with Levene’s test or Bartlett’s test
21. What are the different types of ANOVA tests?
 - There are several types of ANOVA (Analysis of Variance) tests, and the one you choose depends on your research design — like how many factors you're testing, how many groups, and whether you're using the same or different subjects in each condition.
 1. One-Way ANOVA
 - Use when: You’re comparing three or more groups based on one independent variable (factor).
 -  Example: Compare test scores across 3 teaching methods (lecture, group work, online).
 -  Answers: “Do the means differ across these groups?”
 2. Two-Way ANOVA (Factorial ANOVA)
 - Use when: You have two independent variables, and you want to study:
 - The individual effect of each factor
 - The interaction between them
 - Example: Study exam scores based on teaching method (3 levels) and study time (short vs long).
 3. Repeated Measures ANOVA
 - Use when: The same subjects are measured multiple times (like before/after testing or under different conditions).
 - Example: Test heart rate of the same group before, during, and after exercise.
 - This controls for individual differences and is more powerful than one-way ANOVA in repeated scenarios.
 4. Mixed-Design ANOVA (Split-Plot ANOVA)
 - Use when: You have both between-subjects and within-subjects factors.
 - Example: Compare memory test scores for different age groups (between-subjects), tested across three time points (within-subjects).
 - This combines features of two-way and repeated measures ANOVA.
 5. MANOVA (Multivariate ANOVA)
 - Use when: You have two or more dependent variables you want to analyze simultaneously.
 -  Example: Compare how three diet types affect both weight loss and cholesterol.
 - Helps avoid inflating Type I error by testing outcomes together.

```python
2. #Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define parameters for the discrete uniform distribution
low = 1       # Minimum value
high = 6      # Maximum value (inclusive)

# Generate the distribution
dist = randint(low, high + 1)  # +1 because `high` is exclusive in `randint`

# Generate all possible outcomes
x = np.arange(low, high + 1)  # [1, 2, 3, 4, 5, 6]

# Calculate PMF (Probability Mass Function)
pmf = dist.pmf(x)

# Plot the PMF
plt.figure(figsize=(8, 5))
plt.bar(x, pmf, color='skyblue', edgecolor='black', width=0.6)

# Customize the plot
plt.title("Discrete Uniform Distribution PMF (1 to 6)", fontsize=14)
plt.xlabel("Outcome (k)", fontsize=12)
plt.ylabel("Probability P(X=k)", fontsize=12)
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Display the plot
plt.show()
```

```python
3. # Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
import numpy as np
import matplotlib.pyplot as plt

def bernoulli_pdf(p):
    """
    Calculate and plot the PDF of a Bernoulli distribution.

    Parameters:
        p (float): Probability of success (1), where 0 <= p <= 1.

    Returns:
        dict: PDF values for k=0 and k=1.
    """
    if not 0 <= p <= 1:
        raise ValueError("Probability p must be between 0 and 1.")

    pdf = {0: 1 - p, 1: p}  # PDF values

    # Plotting
    plt.figure(figsize=(8, 5))
    plt.bar(pdf.keys(), pdf.values(), color=['red', 'green'], edgecolor='black', width=0.4)
    plt.title(f"Bernoulli Distribution PDF (p={p})", fontsize=14)
    plt.xlabel("Outcome (k)", fontsize=12)
    plt.ylabel("Probability P(X=k)", fontsize=12)
    plt.xticks([0, 1], labels=["Failure (0)", "Success (1)"])
    plt.ylim(0, 1)
    plt.grid(axis='y', linestyle='--', alpha=0.6)

    return pdf

# Example usage
p = 0.7  # Probability of success
pdf = bernoulli_pdf(p)
plt.show()
```

```python
6. #Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
#uniform distribution.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range of the discrete uniform distribution
low = 1  # inclusive lower bound
high = 7 # exclusive upper bound (i.e., values from 1 to 6)

# Create discrete uniform distribution
dist = randint(low, high)

# Define the x values (support of the distribution)
x = np.arange(low - 1, high + 1)  # extend one value below and above for step-like CDF
cdf = dist.cdf(x)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(x, cdf, where='post', label='CDF', color='blue')
plt.scatter(x, cdf, color='red')  # optional: dots on each step
plt.title('CDF of Discrete Uniform Distribution (randint({}, {}))'.format(low, high - 1))
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.xticks(x)
plt.grid(True)
plt.legend()
plt.show()
```

```python
9. # Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores for a dataset and plot them.

    Parameters:
    - data: list or numpy array of numeric values
    """
    # Convert to NumPy array
    data = np.array(data)

    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(10, 5))
    plt.plot(z_scores, marker='o', linestyle='-', color='purple')
    plt.axhline(y=0, color='black', linestyle='--', linewidth=1)
    plt.title("Z-scores of the Dataset")
    plt.xlabel("Data Point Index")
    plt.ylabel("Z-score")
    plt.grid(True)
    plt.show()

    return z_scores
```

```python
16. # Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of a dataset.
    """
    n = len(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation
    z_score = norm.ppf(1 - (1 - confidence) / 2)

    margin_of_error = z_score * (std / np.sqrt(n))
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    print(f"Sample Mean: {mean:.2f}")
    print(f"Confidence Level: {confidence*100:.0f}%")
    print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    return (lower_bound, upper_bound)
data = [68, 72, 70, 74, 69, 71, 73, 75, 70, 68]

# Calculate and interpret CI
confidence_interval(data, confidence=0.95)
```

```python
3. #  Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, pop_mean, pop_std, alpha=0.05):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    # Z-test statistic
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z)))  # two-tailed p-value

    # Output
    print("=== One-Sample Z-Test ===")
    print(f"Sample Mean       : {sample_mean:.2f}")
    print(f"Population Mean   : {pop_mean}")
    print(f"Z-score           : {z:.4f}")
    print(f"P-value           : {p_value:.4f}")

    # Hypothesis test result
    if p_value < alpha:
        print(f"Result: Reject the null hypothesis (α = {alpha})")
    else:
        print(f"Result: Fail to reject the null hypothesis (α = {alpha})")
    return z, p_value

# Example Usage
np.random.seed(42)
sample = np.random.normal(loc=102, scale=10, size=50)  # Simulated sample data
population_mean = 100
population_std = 10  # Known population std

one_sample_z_test(sample, population_mean, population_std)
```

```python
4. #Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z)))  # two-tailed

    print("=== Two-Tailed Z-Test ===")
    print(f"Sample Mean       : {sample_mean:.2f}")
    print(f"Z-score           : {z:.4f}")
    print(f"P-value           : {p_value:.4f}")

    # Critical z for two-tailed test
    z_critical = norm.ppf(1 - alpha/2)

    if abs(z) > z_critical:
        result = f"Reject H₀ (Z = {z:.2f} > {z_critical:.2f})"
    else:
        result = f"Fail to reject H₀ (Z = {z:.2f} ≤ {z_critical:.2f})"

    print("Result:", result)

    # Visualization
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Standard Normal Distribution', color='blue')
    plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.3, label='Rejection Region')
    plt.axvline(z, color='green', linestyle='--', linewidth=2, label=f'Z-score = {z:.2f}')
    plt.axvline(-z_critical, color='black', linestyle='--', label=f'Critical Z = ±{z_critical:.2f}')
    plt.axvline(z_critical, color='black', linestyle='--')

    plt.title("Two-Tailed Z-Test: Decision Regions")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    return z, p_value

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=102, scale=10, size=50)
population_mean = 100
population_std = 10

two_tailed_z_test(sample_data, population_mean, population_std)
```

```python
5. # Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu0=100, mu1=105, sigma=10, n=50, alpha=0.05):
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical values under H0
    z_critical = norm.ppf(1 - alpha/2)
    lower_crit = mu0 - z_critical * se
    upper_crit = mu0 + z_critical * se

    # X-axis range
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # H0 and H1 distributions
    y0 = norm.pdf(x, mu0, se)  # Null hypothesis distribution
    y1 = norm.pdf(x, mu1, se)  # Alternative hypothesis distribution

    # Plot
    plt.figure(figsize=(12, 6))
    plt.plot(x, y0, label='H₀: μ = {}'.format(mu0), color='blue')
    plt.plot(x, y1, label='H₁: μ = {}'.format(mu1), color='green')

    # Shade Type I Error (α) regions
    plt.fill_between(x, y0, where=(x < lower_crit) | (x > upper_crit), color='red', alpha=0.3, label='Type I Error (α)')

    # Shade Type II Error (β) region
    plt.fill_between(x, y1, where=(x >= lower_crit) & (x <= upper_crit), color='orange', alpha=0.3, label='Type II Error (β)')

    # Labels and legend
    plt.axvline(lower_crit, color='black', linestyle='--', label=f'Lower Critical = {lower_crit:.2f}')
    plt.axvline(upper_crit, color='black', linestyle='--', label=f'Upper Critical = {upper_crit:.2f}')
    plt.title("Type I and Type II Errors Visualization")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Run the function
visualize_errors()
```

```python
6. #  Write a Python program to perform an independent T-test and interpret the results.
import numpy as np
from scipy.stats import ttest_ind

def independent_t_test(group1, group2, alpha=0.05):
    # Perform independent t-test
    t_stat, p_value = ttest_ind(group1, group2, equal_var=False)  # Welch’s T-test for unequal variances

    print("=== Independent T-Test ===")
    print(f"Group 1 Mean: {np.mean(group1):.2f}")
    print(f"Group 2 Mean: {np.mean(group2):.2f}")
    print(f"T-statistic : {t_stat:.4f}")
    print(f"P-value     : {p_value:.4f}")

    # Interpret result
    if p_value < alpha:
        print(f"Result: Reject the null hypothesis (means are significantly different at α = {alpha})")
    else:
        print(f"Result: Fail to reject the null hypothesis (no significant difference at α = {alpha})")

    return t_stat, p_value

# Example usage
np.random.seed(42)
group1 = np.random.normal(loc=100, scale=10, size=30)
group2 = np.random.normal(loc=105, scale=10, size=30)

independent_t_test(group1, group2)
```

```python
7. # Perform a paired sample T-test using Python and visualize the comparison results
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

def paired_t_test_visual(before, after, alpha=0.05):
    # Calculate differences
    differences = after - before

    # Paired T-test
    t_stat, p_value = ttest_rel(before, after)

    print("=== Paired Sample T-Test ===")
    print(f"Mean Before: {np.mean(before):.2f}")
    print(f"Mean After : {np.mean(after):.2f}")
    print(f"Mean Difference: {np.mean(differences):.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value    : {p_value:.4f}")

    # Interpret the result
    if p_value < alpha:
        print("Result: Reject the null hypothesis — there is a significant difference.")
    else:
        print("Result: Fail to reject the null hypothesis — no significant difference.")

    # Visualization
    plt.figure(figsize=(10, 5))
    index = np.arange(len(before))
    width = 0.35

    plt.bar(index, before, width, label='Before', alpha=0.7)
    plt.bar(index + width, after, width, label='After', alpha=0.7)

    plt.xlabel('Subjects')
    plt.ylabel('Scores')
    plt.title('Paired Sample T-Test: Before vs After')
    plt.xticks(index + width / 2, [f'#{i+1}' for i in index], rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.grid(True, linestyle='--', alpha=0.4)
    plt.show()

# Example usage
np.random.seed(42)
before_scores = np.random.normal(loc=75, scale=5, size=20)
after_scores = before_scores + np.random.normal(loc=2, scale=3, size=20)

paired_t_test_visual(before_scores, after_scores)
```

```python
9. #Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)

    # t-critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    # margin of error
    margin = t_crit * std_err

    # confidence interval
    lower = mean - margin
    upper = mean + margin

    print(f"Sample Mean      : {mean:.2f}")
    print(f"Confidence Level : {confidence*100:.0f}%")
    print(f"Confidence Interval: ({lower:.2f}, {upper:.2f})")

    return (lower, upper)
```

```python
10. # Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    std_err = sample_std / np.sqrt(n)

    # t-critical value for two-tailed test
    t_crit = t.ppf((1 + confidence) / 2, df=n - 1)

    # Margin of error
    margin = t_crit * std_err

    print(f"Sample Mean      : {sample_mean:.2f}")
    print(f"Sample Size (n)  : {n}")
    print(f"Sample Std Dev   : {sample_std:.2f}")
    print(f"Confidence Level : {confidence*100:.0f}%")
    print(f"Margin of Error  : ±{margin:.2f}")

    return margin
```

```python
11. #  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayesian_inference(prior, likelihood, false_positive):
    # Total probability of testing positive (Evidence)
    evidence = (likelihood * prior) + (false_positive * (1 - prior))

    # Posterior probability
    posterior = (likelihood * prior) / evidence

    print("=== Bayesian Inference ===")
    print(f"Prior (P(Disease))         : {prior:.4f}")
    print(f"Likelihood (P(Pos|Disease)): {likelihood:.4f}")
    print(f"False Positive Rate        : {false_positive:.4f}")
    print(f"Evidence (P(Pos))          : {evidence:.4f}")
    print(f"Posterior (P(Disease|Pos)) : {posterior:.4f}")

    return posterior
```

```python
16. # Implement an F-test using Python to compare the variances of two random samples.
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    # Sample sizes
    n1 = len(sample1)
    n2 = len(sample2)

    # Sample variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Ensure F >= 1 by setting the larger variance as numerator
    if var1 > var2:
        f_stat = var1 / var2
        dfn = n1 - 1
        dfd = n2 - 1
    else:
        f_stat = var2 / var1
        dfn = n2 - 1
        dfd = n1 - 1

    # Calculate critical value and p-value
    p_value = 1 - f.cdf(f_stat, dfn, dfd)
    critical_value = f.ppf(1 - alpha, dfn, dfd)

    # Output
    print("=== F-Test for Equal Variances ===")
    print(f"Variance 1: {var1:.4f}, Variance 2: {var2:.4f}")
    print(f"F-statistic: {f_stat:.4f}")
    print(f"Degrees of Freedom: ({dfn}, {dfd})")
    print(f"Critical value (α={alpha}): {critical_value:.4f}")
    print(f"P-value: {p_value:.4f}")

    if f_stat > critical_value:
        print("Result: Reject the null hypothesis — variances are significantly different.")
    else:
        print("Result: Fail to reject the null hypothesis — no significant difference in variances.")
```

```python
24. # Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# Simulate sample data
np.random.seed(42)
sample = np.random.normal(loc=100, scale=15, size=30)  # mean=100, std=15

# Known or hypothesized population variance
sigma0_squared = 225  # σ₀² = 15²

# Sample statistics
n = len(sample)
sample_variance = np.var(sample, ddof=1)  # unbiased sample variance

# Chi-square statistic
chi_square_stat = (n - 1) * sample_variance / sigma0_squared

# Degrees of freedom
df = n - 1

# P-value for two-tailed test
p_value = 2 * min(chi2.cdf(chi_square_stat, df), 1 - chi2.cdf(chi_square_stat, df))

# Print results
print("=== Chi-square Test for Population Variance ===")
print(f"Sample Variance: {sample_variance:.4f}")
print(f"Chi-square Statistic: {chi_square_stat:.4f}")
print(f"Degrees of Freedom: {df}")
print(f"P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Result: Reject the null hypothesis (variance differs significantly).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference in variance).")

# Visualization
x = np.linspace(chi2.ppf(0.001, df), chi2.ppf(0.999, df), 500)
y = chi2.pdf(x, df)

plt.figure(figsize=(10, 5))
plt.plot(x, y, 'b-', label='Chi-square PDF')
plt.axvline(chi_square_stat, color='red', linestyle='--', label='Test Statistic')
plt.fill_between(x, 0, y, where=((x < chi2.ppf(0.025, df)) | (x > chi2.ppf(0.975, df))),
                 color='red', alpha=0.3, label='Rejection Region (α = 0.05)')
plt.title('Chi-square Test for Population Variance')
plt.xlabel('Chi-square value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
25. #Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.
import numpy as np
from scipy.stats import norm

def z_test_proportions(x1, n1, x2, n2, alpha=0.05):
    # Proportions
    p1 = x1 / n1
    p2 = x2 / n2

    # Pooled proportion
    p_pool = (x1 + x2) / (n1 + n2)

    # Z-score
    z = (p1 - p2) / np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))

    # P-value (two-tailed)
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Critical value
    z_crit = norm.ppf(1 - alpha / 2)

    # Output
    print("=== Z-Test for Comparing Proportions ===")
    print(f"Group 1: {x1}/{n1} = {p1:.4f}")
    print(f"Group 2: {x2}/{n2} = {p2:.4f}")
    print(f"Z-statistic: {z:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Critical Z-value (±{z_crit:.2f})")

    if abs(z) > z_crit:
        print("Result: Reject the null hypothesis (significant difference in proportions).")
    else:
        print("Result: Fail to reject the null hypothesis (no significant difference).")

# Example
z_test_proportions(x1=45, n1=100, x2=30, n2=100)
```

```python
26. # Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.
import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt

# Generate two datasets with different variances
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=50)
data2 = np.random.normal(loc=50, scale=5, size=50)

# Sample sizes
n1, n2 = len(data1), len(data2)

# Sample variances
s1_sq = np.var(data1, ddof=1)
s2_sq = np.var(data2, ddof=1)

# Assign larger variance to numerator to keep F ≥ 1
if s1_sq >= s2_sq:
    F = s1_sq / s2_sq
    df1, df2 = n1 - 1, n2 - 1
else:
    F = s2_sq / s1_sq
    df1, df2 = n2 - 1, n1 - 1

# Two-tailed p-value
p_value = 2 * min(f.cdf(F, df1, df2), 1 - f.cdf(F, df1, df2))

# Critical F-values for 95% confidence
alpha = 0.05
f_crit_low = f.ppf(alpha / 2, df1, df2)
f_crit_high = f.ppf(1 - alpha / 2, df1, df2)

# Print results
print("=== F-Test for Variance Comparison ===")
print(f"Variance 1: {s1_sq:.4f}")
print(f"Variance 2: {s2_sq:.4f}")
print(f"F-statistic: {F:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Critical values: ({f_crit_low:.4f}, {f_crit_high:.4f})")

if F < f_crit_low or F > f_crit_high:
    print("Result: Reject the null hypothesis (variances are significantly different).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")

# Visualization
x = np.linspace(f.ppf(0.001, df1, df2), f.ppf(0.999, df1, df2), 500)
y = f.pdf(x, df1, df2)

plt.figure(figsize=(10, 5))
plt.plot(x, y, 'b-', lw=2, label='F-distribution PDF')
plt.axvline(F, color='red', linestyle='--', label='F-statistic')
plt.axvline(f_crit_low, color='gray', linestyle='--', label='Critical values')
plt.axvline(f_crit_high, color='gray', linestyle='--')
plt.fill_between(x, 0, y, where=(x < f_crit_low) | (x > f_crit_high), color='red', alpha=0.3, label='Rejection Region')
plt.title('F-Test for Comparing Two Variances')
plt.xlabel('F value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
27. #Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Simulate observed frequencies (e.g., from rolling a die)
np.random.seed(42)
observed = np.random.multinomial(n=120, pvals=[1/6]*6)
expected = [120 / 6] * 6  # uniform expected frequency

# Perform Chi-square goodness of fit test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Output results
print("=== Chi-Square Goodness of Fit Test ===")
print(f"Observed Frequencies: {observed}")
print(f"Expected Frequencies: {expected}")
print(f"Chi-square Statistic: {chi_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Result: Reject the null hypothesis (the distribution differs from expected).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")

# Visualization
labels = ['1', '2', '3', '4', '5', '6']
x = np.arange(len(labels))
width = 0.35

plt.figure(figsize=(8, 5))
plt.bar(x - width/2, observed, width, label='Observed', color='skyblue')
plt.bar(x + width/2, expected, width, label='Expected', color='orange')
plt.xticks(x, labels)
plt.ylabel('Frequency')
plt.title('Observed vs Expected Frequencies')
plt.legend()
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
